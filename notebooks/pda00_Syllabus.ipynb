# 파이썬 데이터분석 강좌 소개

## 목표
* 파이썬 프로그래밍 기초 습득
* 데이터 과학 및 머신러닝/딥러닝 기초다지기

## 내용
* 파이썬 프로그래밍 기초와 머신러닝/딥러닝 기초 이론 및 실습

## 인공지능, 데이터 과학, 머신러닝, 딥러닝의 관계
* 인공지능: 사고나 학습 등 인간이 가진 지적능력을 컴퓨터를 통해 구현하는 기술
* 데이터 과학: 수학과 통계 지식을 활용하여 데이터로부터 특정 정보를 추출하는 과학
* 머신러닝: 컴퓨터가 데이터를 스스로 학습하여 인공지능의 성능을 향상시키는 기법
* 딥러닝: 인공신경망 이론을 기반으로 복잡한 비선형 문제를 기계 스스로 학습하도록 하는 기법

## 데이터 과학, 머신러닝/딥러닝을 위한 필수 요소
* 수학 및 통계: 선형대수학, 확률과 통계, 미적분학
* 컴퓨터 프로그래밍, 특히 파이썬 프로그래밍 기초

## 전제조건

* 기본적으로 없음. 파이썬 프로그래밍이 처음이라는 가정하에 수업을 진행할 것임.

## 교재

* '밑바닥부터 시작하는 데이터 과학', 조엘 그루스, 인사이트
